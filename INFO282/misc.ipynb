```python
from mathesis.grammars import BasicGrammar
from mathesis.semantics.truth_table import ClassicalTruthTable
import itertools

comb = ["(p → (q ∧ r))",
        "(p ∧ (q → r))", 
        "((p ∧ q) → r)",
        "(p → (q → r))",
]

grammer = BasicGrammar()

for pair in itertools.permutations(comb, 2):
    string = f"{pair[0]} → {pair[1]}"
    fml = grammer.parse(string)
    table = ClassicalTruthTable(fml)
    print(f"{string}: {table.is_valid()}")


fml = grammer.parse("((p ∧ q) → r) → (p → (q → r))")
table = ClassicalTruthTable(fml)
table
```
